```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
from tensorflow.keras.models import Sequential

model = Sequential()
print(model)
```

```python
from tensorflow.keras.layers import Dense

model.add(Dense(units=4, input_shape=(10, )))
model.summary()
```

```python
model.add(Dense(units=2))
model.summary()
```

```python
from tensorflow.keras.activations import linear     #funkcje aktywacji

random_data = sorted(np.random.randn(200))
data = pd.DataFrame({'data': random_data, 'linear': linear(random_data)})

px.line(data, x='data', y='linear', width=800)
```

```python
from tensorflow.keras.activations import sigmoid     #funkcje aktywacji

random_data = sorted(np.random.randn(200))
data = pd.DataFrame({'data': random_data, 'sigmoid': sigmoid(random_data)})

px.line(data, x='data', y='sigmoid', width=800)
```

```python
from tensorflow.keras.activations import relu     #funkcje aktywacji

random_data = sorted(np.random.randn(200))
data = pd.DataFrame({'data': random_data, 'relu': relu(random_data)})

px.line(data, x='data', y='relu', width=800)
```

```python
from tensorflow.keras.activations import tanh     #funkcje aktywacji

random_data = sorted(np.random.randn(200))
data = pd.DataFrame({'data': random_data, 'tanh': tanh(random_data)})

px.line(data, x='data', y='tanh', width=800)
```

```python
#budowa modelu

model = Sequential()
model.add(Dense(units=8, activation='relu', input_shape=(10, )))
model.add(Dense(units=1, activation='sigmoid'))
model.summary()
```

```python
#klasyfikacja binarna
model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

#klasyfikacja wieloklasowa
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

#regresja
model.compile(optimizer='rmsprop',
              loss='mae')
```
