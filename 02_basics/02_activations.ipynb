```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.graph_objects as go
import plotly.express as px
import math

np.set_printoptions(precision=6)
```

```python
#ReLU
def relu(x):
    return max(0, x)

data = 2 * np.random.randn(1000)
data = sorted(data)
data_relu = np.array([relu(i) for i in data])

df = pd.DataFrame({'data': data, 'data_relu': data_relu})

px.line(df, x='data', y='data_relu', width=700, height=400, title='ReLU function')
```

```python
#Sigmoid
def sigmoid(x):
    beta = 1
    return 1 / (1 + np.exp(-x*beta))

data = 2 * np.random.randn(1000)
data = sorted(data)
y = np.array([sigmoid(i) for i in data])

df = pd.DataFrame({'data': data, 'y': y})

px.line(df, x='data', y='y', width=700, height=400, title='Sigmoid function')
```

```python
#Tanh
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

data = 2 * np.random.randn(1000)
data = sorted(data)
y = np.array([tanh(i) for i in data])

df = pd.DataFrame({'data': data, 'y':y})

px.line(df, x='data', y='y', width=700, height=400, title='Tanh function')
```
